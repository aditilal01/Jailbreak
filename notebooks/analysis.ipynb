# Say It Differently — Results Analysis

Interactive analysis of how linguistic style variations affect LLM jailbreak success rates.

**Paper**: *"Say It Differently: Linguistic Styles as Jailbreak Vectors"*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

sns.set_theme(style='whitegrid', font_scale=1.2)
print('Libraries loaded successfully.')
```

## 1. Load Results

```python
# Load all results
df = pd.read_csv('../results/all_results.csv')
print(f'Total evaluations: {len(df)}')
print(f'Models: {df["model"].unique().tolist()}')
print(f'Styles: {df["style"].unique().tolist()}')
df.head()
```

## 2. ASR by Model

```python
asr_model = pd.read_csv('../results/asr_by_model.csv', index_col=0)
asr_model['asr_pct'] = asr_model['asr'] * 100
print(asr_model.to_string())

fig, ax = plt.subplots(figsize=(10, 5))
asr_model['asr_pct'].plot(kind='bar', color=sns.color_palette('viridis', len(asr_model)), ax=ax)
ax.set_ylabel('ASR (%)')
ax.set_title('Attack Success Rate by Model')
ax.set_ylim(0, 100)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

## 3. ASR by Style

```python
asr_style = pd.read_csv('../results/asr_by_style.csv', index_col=0)
asr_style['asr_pct'] = asr_style['asr'] * 100
print(asr_style.to_string())

fig, ax = plt.subplots(figsize=(10, 7))
colors = sns.color_palette('RdYlGn_r', len(asr_style))
asr_style['asr_pct'].plot(kind='barh', color=colors, ax=ax)
ax.set_xlabel('ASR (%)')
ax.set_title('Attack Success Rate by Linguistic Style')
ax.set_xlim(0, 100)
plt.tight_layout()
plt.show()
```

## 4. Heatmap: Style × Model

```python
pivot = df.pivot_table(values='is_unethical', index='style', columns='model', aggfunc='mean') * 100

fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(pivot, annot=True, fmt='.1f', cmap='YlOrRd', linewidths=0.5, ax=ax,
            cbar_kws={'label': 'ASR (%)'})
ax.set_title('ASR Heatmap: Linguistic Style × Target Model')
plt.tight_layout()
plt.show()
```

## 5. Neutral vs. Stylized Comparison

```python
comparison = pd.read_csv('../results/neutral_vs_stylized.csv')
print('Neutral vs. Stylized Comparison:')
print(f"  Neutral ASR:           {comparison['neutral_asr'].values[0]*100:.1f}%")
print(f"  Stylized ASR:          {comparison['stylized_asr'].values[0]*100:.1f}%")
print(f"  Absolute increase:     +{comparison['asr_increase'].values[0]*100:.1f}%")
print(f"  Relative increase:     {comparison['relative_increase_pct'].values[0]:.1f}%")
```
